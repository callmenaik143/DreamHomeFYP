# API

```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image

home_model_path = "/content/drive/MyDrive/1flooring_stable-diffusion-v1.4"
home_pipe = StableDiffusionPipeline.from_pretrained(home_model_path, torch_dtype=torch.float16)

home_pipe.to("cuda")

```

```python
!npm install -g localtunnel
```

```python
from flask import Flask, request, send_file
from PIL import Image
import os
import threading
app = Flask(__name__)

def generate_home(prompt):
    image = home_pipe(prompt=prompt).images[0]
    image_path = '/content/generated_home_image.png'
    image.save(image_path)
    return image_path

@app.route('/generate-home', methods=['POST'])
def generate_home_api():
    # Extract parameters from the request
    data = request.json
    prompt = data.get('query')
    # Generate the image based on the parameters
    image_path = generate_home(prompt)
    # Return the generated image file
    return send_file(image_path, mimetype='image/png')

def run_flask():
    app.run(host='0.0.0.0', port=8050)
flask_thread = threading.Thread(target=run_flask)
flask_thread.start()
```
